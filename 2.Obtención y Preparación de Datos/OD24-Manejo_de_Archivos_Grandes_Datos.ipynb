## <font color='blue'>__Trabajar con grandes datos (*Big Data*)__</font>
Si sus archivos son demasiado grandes para guardarlos o procesarlos, hay varios enfoques que puede tomar para reducir el espacio en disco requerido:

* Comprimir los archivos
* Elegir solo las columnas que necesitemos
* Omitir las filas que no necesitamos
* Forzar el uso de tipos de datos menos precisos
* Dividir los datos en trozos

Daremos un vistazo cada una de estas técnicas.

### __Comprimir y Descomprimir Archivos__
Puede crear un archivo de almacenamiento como lo haría con uno normal, con la adición de un sufijo que corresponda al tipo de compresión deseado:

* `'.gz'`
* `'.bz2'`
* `'.zip'
`* `'.xz'`

Pandas puede deducir el tipo de compresión por sí mismo:

````
df = pd.DataFrame(data=data).T
df.to_csv('data.csv.zip')
````

````
!ls -l data.cs*
````

Vemos que se creó un archivo .csv comprimido como un nuevo archivo. El tamaño del archivo .csv normal es de 1031 bytes, mientras que el archivo comprimido solo tiene 733 bytes.

Puede abrir este archivo comprimido como de costumbre con la función pandas `read_csv()`:

```python
df = pd.read_csv('data.csv.zip', index_col=0,
                 parse_dates=['IND'])
```

`read_csv()` descomprime el archivo antes de leerlo en un DataFrame.

Puede especificar el tipo de compresión con el parámetro opcional `compression` , el cual puede tomar cualquiera de los siguientes valores:

* `'inferir'`
* `'gzip'`
* `'bz2'`
* `'zip'`
* `'xz'`
* `None`

El valor predeterminado `compression='infer'` indica que Pandas debe deducir el tipo de compresión de la extensión del archivo.

Así es como comprimiría un archivo pickle:



````
df = pd.DataFrame(data=data).T
df.to_pickle('data.pickle.compress', compression='gzip')
````

### __Elegir columnas__
Las funciones `read_csv()` y `read_excel()` de Pandas tienen el parámetro opcional `usecols` que puede usar para especificar las columnas que desea cargar desde el archivo. Puede pasar la lista de nombres de columnas como el argumento correspondiente:

````
df = pd.DataFrame(data=data).T
df.to_csv('data.csv')
````

````
df = pd.read_csv('data.csv', usecols=['PAIS', 'AREA'])
df
````

En lugar de los nombres de las columnas, también puede pasar sus índices:

````
df = pd.read_csv('data.csv',index_col=0, usecols=[0, 1, 3])
df
````

### __Omitir filas__
Cuando estamos desarrollando nuestros algoritmos para procesar datos o para el aprendizaje automático, a menudo no necesitaremos el conjunto de datos completo. Es conveniente cargar solo un subconjunto de los datos para acelerar el proceso. Las funciones `read_csv()` y `read_excel()` de Pandas tienen algunos parámetros opcionales que le permiten seleccionar qué filas desea cargar:

* `skiprows`: número de filas para omitir al principio del archivo si es un número entero, o los índices de base cero de las filas para omitir si es un objeto similar a una lista
* `skipfooter`: el número de filas a saltar al final del archivo
* `nrows`: el número de filas para leer

Así es como omitiría filas con índices impares basados en cero, manteniendo los pares:

````
df = pd.read_csv('data.csv', index_col=0, skiprows=range(1, 20, 2))
df
````

### __Forzar tipos de datos menos precisos__
Utilizar tipos de datos menos precisos puede ahorrar una cantidad significativa de memoria. Primero, obtenga los tipos de datos con `.dtypes`.

Las columnas con los números de punto flotante son flotantes de 64 bits. Cada número de este tipo `float64` consume 64 bits u 8 bytes. Cada columna tiene 21 números y requiere 168 bytes. Puede verificar esto con `.memory_usage()`:

````
df = pd.read_csv('data.csv', index_col=0, parse_dates=['IND'])
print(df.dtypes, '\n\n', 'Uso de memoria')
df.memory_usage()
````

Si re-definimos los tipos de datos numéricos a `float32`, reduciremos el espacio utilizado. Esto puede ser relevante si nustros archivos son muy grandes. Debemos recordar que Pandas funciona con sus archivos en memoria. Además de ahorrar memoria, puede reducir significativamente el tiempo necesario para procesar datos utilizando `float32` en lugar de `float64` en algunos casos.

````
# Los tipos los pasamos como un diccionario
dtypes = {'POB': 'float32', 'AREA': 'float32', 'PGB': 'float32'}
df = pd.read_csv('data.csv', index_col=0, dtype=dtypes,
                 parse_dates=['IND'])
print(df.dtypes, '\n\n', 'Uso de memoria')
df.memory_usage()
````

A veces, el utilizar datos categóricos también ayuda. Observe que modificamos _OCDE_ desde el tipo `object`, inferido por defecto por Pandas, a `category`. No ocurre lo mismo con _CONT_.

````
# Modificamos el dtype de 'OCDE'
dtypes = {'POB': 'float32', 'AREA': 'float32', 'PGB': 'float32',
          'CONT': 'category', 'OCDE': 'category'}
df = pd.read_csv('data.csv', index_col=0, dtype=dtypes,
                 parse_dates=['IND'])
print(df.dtypes, '\n\n', 'Uso de memoria')
df.memory_usage()
````

### __Utilizando trozos (_chunks_) de archivos e iterar sobre ellos__
Otra forma de manejar conjuntos de datos muy grandes es dividir los datos en partes más pequeñas y procesar una parte a la vez. Si usa `read_csv()`, `read_json()` o `read_sql()`, puede especificar el parámetro opcional `chunksize`:

````
data_chunk = pd.read_csv('data.csv', index_col=0, chunksize=8)
print(type(data_chunk))

print(hasattr(data_chunk, '__iter__'))

print(hasattr(data_chunk, '__next__'))
````

El valor predeterminado de `chunksize` es `None`. Puede tomar un valor entero que indica la cantidad de elementos en un solo fragmento. Cuando `chunksize` es un número entero, `read_csv()` devuelve un iterable que puede usar en un bucle for para obtener y procesar solo un fragmento del conjunto de datos en cada iteración:

````
for df_chunk in pd.read_csv('data.csv', index_col=0, chunksize=4):
    print(df_chunk, end='\n\n')
    print('Memoria:', df_chunk.memory_usage().sum(), 'bytes',
          end='\n\n\n')
````

En este ejemplo, el tamaño de fragmento es 4. La primera iteración del ciclo `for` devuelve un DataFrame con las primeras cuatro filas del conjunto de datos únicamente. La segunda iteración devuelve otro DataFrame con las siguientes cuatro filas. Así sucesivamente, hasta que el últmo ciclo devuelve las filas restantes, en este caso una.
